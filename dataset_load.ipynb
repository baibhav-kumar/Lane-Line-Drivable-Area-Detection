```python
# ===============================
# ğŸ“¥ DOWNLOAD & UNZIP BDD100K DATASET FROM GOOGLE DRIVE
# ===============================

import os
import zipfile
import subprocess
import gdown

# --- 1ï¸âƒ£ Ensure gdown is installed ---
try:
    import gdown
except ImportError:
    subprocess.check_call(["pip", "install", "gdown"])

# --- 2ï¸âƒ£ Define paths ---
project_dir = os.path.expanduser("~/iotlab_project/TwinLiteNet")
zip_path = os.path.join(project_dir, "bdd100k.zip")
extract_dir = os.path.join(project_dir, "data", "bdd100k")

# Google Drive link (convert to direct download format)
gdrive_url = "https://drive.google.com/uc?id=1Y-00ftXTA3MC548M3Lm4eTV7Du3SxMOL"

# --- 3ï¸âƒ£ Make sure destination folder exists ---
os.makedirs(extract_dir, exist_ok=True)
print(f"ğŸ“‚ Using folder: {extract_dir}")

# --- 4ï¸âƒ£ Download ZIP file if it doesnâ€™t already exist ---
if not os.path.exists(zip_path):
    print("â¬‡ï¸ Downloading BDD100K dataset from Google Drive...")
    gdown.download(gdrive_url, zip_path, quiet=False)
else:
    print(f"âœ… ZIP file already exists at: {zip_path}")

# --- 5ï¸âƒ£ Extract ZIP file ---
print("ğŸ“¦ Extracting files...")
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)
print(f"âœ… Extracted to: {extract_dir}")

# --- 6ï¸âƒ£ Verify extraction ---
print("\nğŸ“ Sample of extracted files:")
for root, dirs, files in os.walk(extract_dir):
    for f in files[:10]:  # show a few files
        print(os.path.join(root, f))
    break
```
